```python

import numpy as np
import random

num_cities = 30
distances = np.random.randint(1, 20, size=(num_cities, num_cities))
np.fill_diagonal(distances, 0)
distances = (distances + distances.T) // 2

alpha = 1
beta = 3
evaporation_rate = 0.8
num_ants = 10
num_iterations = 60

# INITIALIZE PHEROMONE MATRIX
#Start with equal pheromone levels on all edges
pheromones = np.ones((num_cities, num_cities))


#CALCULATE TRANSITION PROBABILITIES
def calculate_probabilities(current_city, unvisited):
    probabilities = []
    for city in unvisited:
        #Pheromone influence
        tau = pheromones[current_city][city] ** alpha
        #Distance heuristic
        eta = (1 / distances[current_city][city]) ** beta
        #Combined desirability
        probabilities.append(tau * eta)
    probabilities = np.array(probabilities)
    #Normalize probabilities
    probabilities = probabilities / probabilities.sum()
    return probabilities



#BUILD ROUTE FOR ONE ANT
def build_route():
    current_city = random.randint(0, num_cities - 1)
    route = [current_city]
    #cities no visited
    unvisited = list(range(num_cities))
    unvisited.remove(current_city)
    #Continue until all cities are visited
    while unvisited:
        #transition probabilities
        probabilities = calculate_probabilities(current_city, unvisited)
        #Choose the next
        next_city = int(np.random.choice(unvisited, p=probabilities))
        #Add city to route
        route.append(next_city)
        #Update current position
        current_city = next_city
        #Remove visited city
        unvisited.remove(next_city)
    return route


def route_distance(route):
    distance = 0
    # Sum distances between consecutive cities
    for i in range(len(route) - 1):
        distance += distances[route[i]][route[i+1]]
    # Return to starting city
    distance += distances[route[-1]][route[0]]
    return distance


#MAIN LOOP
best_route = None
best_distance = float("inf")
for iteration in range(num_iterations):
    routes = []
    route_distances = []
    #Each ant builds a route
    for ant in range(num_ants):
        route = build_route()
        distance = route_distance(route)
        routes.append(route)
        route_distances.append(distance)
        #Update thw global best
        if distance < best_distance:
            best_distance = distance
            best_route = route

    #EVAPORATION
    pheromones *= (1 - evaporation_rate)
    #DEPOSIT
    for route, distance in zip(routes, route_distances):
        pheromone_deposit = 1 / distance
        for i in range(len(route) - 1):
            a = route[i]
            b = route[i+1]
            #Update pheromone in both directions
            pheromones[a][b] += pheromone_deposit
            pheromones[b][a] += pheromone_deposit




    iteration_best = min(route_distances)

    print(
        "Iteration:", iteration,
        "| Iteration Best:", iteration_best,
        "| Global Best:", best_distance
    )

print("\nBest route found:", best_route)
print("Minimum distance:", best_distance)

```
