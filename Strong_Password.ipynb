```python
import random
import string

PASSWORD_LENGTH = 8
POPULATION_SIZE = 10
GENERATIONS = 50
MUTATION_RATE = 0.1

CHARS = string.ascii_letters + string.digits + "!@#$%"

def create_password():
    return ''.join(random.choice(CHARS) for _ in range(PASSWORD_LENGTH))

#Fitness function
def fitness(password):
    score = 0
    if any(c.islower() for c in password):
        score += 1
    if any(c.isupper() for c in password):
        score += 1
    if any(c.isdigit() for c in password):
        score += 1
    if any(c in "!@#$%" for c in password):
        score += 1
    return score

#Selection
def select(population):
    return sorted(population, key=fitness, reverse=True)[:2]

#Crossover
def crossover(parent1, parent2):
    point = random.randint(1, PASSWORD_LENGTH - 1)
    return parent1[:point] + parent2[point:]

#Mutation
def mutate(password):
    password = list(password)
    for i in range(len(password)):
        if random.random() < MUTATION_RATE:
            password[i] = random.choice(CHARS)
    return ''.join(password)

#Main genetic algorithm
population = [create_password() for _ in range(POPULATION_SIZE)]

for _ in range(GENERATIONS):
    parents = select(population)
    new_population = parents.copy()

    while len(new_population) < POPULATION_SIZE:
        child = crossover(parents[0], parents[1])
        child = mutate(child)
        new_population.append(child)

    population = new_population

best_password = max(population, key=fitness)
print("Best password found:", best_password)
```
