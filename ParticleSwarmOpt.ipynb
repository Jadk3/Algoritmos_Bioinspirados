```python
import numpy as np

class PSO:
    def __init__(self, func, num_particles=10, num_iterations=50):
        self.func = func
        self.num_particles = num_particles
        self.num_iterations = num_iterations

        # PSO parameters
        self.w = 0.5
        self.c1 = 1.5
        self.c2 = 1.5

        # Initialize particles
        self.positions = np.random.uniform(-10, 10, num_particles)
        self.velocities = np.random.uniform(-1, 1, num_particles)

        # Initialize bests
        self.pbest_pos = self.positions.copy()
        self.pbest_val = self.func(self.positions)

        self.gbest_pos = self.pbest_pos[np.argmin(self.pbest_val)]

    def optimize(self):
        for i in range(self.num_iterations):

            r1 = np.random.rand(self.num_particles)
            r2 = np.random.rand(self.num_particles)

            # Update velocities (vectorized)
            self.velocities = (
                self.w * self.velocities
                + self.c1 * r1 * (self.pbest_pos - self.positions)
                + self.c2 * r2 * (self.gbest_pos - self.positions)
            )

            # Update positions
            self.positions += self.velocities

            # Evaluate
            values = self.func(self.positions)

            # Update personal bests
            better_mask = values < self.pbest_val
            self.pbest_pos[better_mask] = self.positions[better_mask]
            self.pbest_val[better_mask] = values[better_mask]

            # Update global best
            self.gbest_pos = self.pbest_pos[np.argmin(self.pbest_val)]

            print(f"Iteration {i+1}: best value = {self.func(self.gbest_pos):.5f}")

        return self.gbest_pos, self.func(self.gbest_pos)


# Objective function
def objective_function(x):
    return x**2


# Run PSO
pso = PSO(objective_function)
best_x, best_val = pso.optimize()

print("\nBest solution found (x):", best_x)
print("Minimum value:", best_val)
```
